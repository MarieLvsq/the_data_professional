```python
sns.catplot("sex", data=titanic, kind="count")
```

# Question: What were the total numbers of passengers in different classes sep by gender?

```python
sns.catplot("class", data=titanic, hue="sex", kind='count')
```

# Question: What were the total numbers of passengers in different classes?

# Question: What were the total numbers of passengers dead or alive?

# Question: What classes had the survivors travelled in?

```python
sns.catplot("class", hue="survived", data=titanic, kind="count")
```

# Question: What classes had the survivors travelled in?

# Question: Which gender on which class survived most?

```python
sns.catplot("class", "survived", hue='sex', data=titanic, kind='point')
```

# Question: Passenger from which port survived most?

```python
sns.catplot("embark_town", "survived", hue='sex', data=titanic, kind='point')
```

# Question: What is the age distribution of passengers?
